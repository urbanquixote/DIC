### ** Key Points of Using Pretrained CNN Models
- 1. 데이터 구축<br>
    - Pretrained model은 사전에 설정한 input size에 적합한 구조를 가지고 있으므로 새로운 데이터를 pretrained model의 input size로 변환해야 함 <br>
- 2. 모델 구축<br>
    - Pretrained model은 ImageNet에 학습되어 1000개의 output node를 가지고 있으므로 해당 모델을 새로운 데이터에 사용하기 위해 pretrained model의 마지막 layer를 새로운 데이터의 class 개수에 맞는 layer로 대체해야 함 <br>
- 3. 모델 학습<br>
    - Pretrained model의 weight를 그대로 사용할 layer와 새로운 데이터에 맞게 업데이트할 layer를 구분하여 optimizer가 업데이트할 weight를 설정해야 함<br><br>

# <br>0. Hyperparameter Setting
- data_dir: 데이터가 존재하는 경로 (해당 실습에서는 새로운 이미지 데이터의 ImageFolder가 존재하는 경로를 의미함)
- model_name: 사용할 pretrained CNN model의 이름으로 ['inception', 'resnet', 'densenet'] 중 하나로 설정 가능함
- batch_size: 학습 및 검증에 사용할 배치의 크기
- num_classes: 새로운 데이터의 class 개수
- num_epochs: 학습할 epoch 횟수
- feature_extract: feature extracting 여부 (True - the reshaped layer params만 학습 / False - the whole model의 params를 finetune)
- random_seed: reproduction을 위해 고정할 seed의 값

```python
# Hyperparameter setting
data_dir = "/content/LG_time_series_day10/input/hymenoptera-data/hymenoptera_data"
model_name = "resnet"
batch_size = 32
num_classes = 2
num_epochs = 10
feature_extract = True

random_seed = 42
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu") # Detect if we have a GPU available
```

```python
# pretrained CNN model의 input size 설정
input_size_dict = {"inception": 299, "resnet": 224, "densenet": 224}
input_size = input_size_dict[model_name]
```

```python
# seed 고정
torch.manual_seed(random_seed)
torch.cuda.manual_seed(random_seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
np.random.seed(random_seed)
random.seed(random_seed)
```

# <br>__1. Data: Hymenoptera Data__
- 데이터 description
    - Hymenoptera Data는 개미와 벌의 이미지로 구성된 데이터셋이다. 해당 데이터셋은 총 398개의 이미지와 해당 이미지에 매칭되는 개미/벌의 label로 구성되어 있으며, 이 중 245개는 train 데이터이고 153개는 test 데이터이다. Hymenoptera Data를 활용한 이미지 분류 task는 이미지를 input으로 받아 이를 개미/벌 중 하나의 class로 분류하는 것을 목표로 한다. <br><br>

    - 독립변수(X): 다양한 크기의 개미/벌 컬러 이미지
    - 종속변수(Y): 이미지의 label - 'ants'(개미) / 'bees'(벌)<br><br>

### Step 1-1. Data Transform 설정하기
- 대표적인 transformation <br>
    - transforms.Resize(input_size): 이미지를 input_size의 크기로 rescale함
    - transforms.CenterCrop(input_size): 이미지의 중심부에서 (input_size x input_size) 크기의 영역을 crop함
    - transforms.ToTensor(): 0-1 사이의 값을 가진 Tensor 형태로 변환함
    - transforms.Normalize(mean, std): mean을 빼고 std로 나누는 정규화를 적용함<br><br>

- 주의) train 데이터에는 아래와 같은 data augmentation을 적용하여 한 이미지로부터 다양한 이미지를 추출하여 학습함으로써 학습의 효과를 높일 수 있음. 하지만, validation 데이터에는 input size 변환 및 normalization만 적용해야 함<br>
    - transforms.RandomResizedCrop(input_size): 이미지의 일부 영역을 랜덤으로 crop한 후, 이를 input_size의 크기로 resize함
    - transforms.RandomHorizontalFlip(): 50%의 확률로 이미지를 좌우 반전함

```python
# input 데이터에 적용할 transform 설정
data_transforms = transforms.Compose([
    transforms.Resize(input_size),
    transforms.CenterCrop(input_size),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

### Step 1-2. Dataset 구축하기

```python
# Dataset 구축 (ImageFolder 활용)
trainset = datasets.ImageFolder(os.path.join(data_dir, 'train'), data_transforms)
testset = datasets.ImageFolder(os.path.join(data_dir, 'val'), data_transforms)
```

### Step 1-3. DataLoader 구축하기 - Train/Validation/Test DataLoader

```python
# train data를 시간순으로 8:2의 비율로 train:validation set으로 분할
# torch.utils.data.sampler.SubsetRandomSampler를 활용하여 간단하게 DataLoader를 구축할 수 있음

# 전체 index의 list와 train 및 validation set을 분할할 기준 설정
indices = list(range(len(trainset)))
split = int(0.8 * len(trainset))

# 전체 index의 list를 섞은 후, train 및 validation set의 index 분할
np.random.seed(random_seed)
np.random.shuffle(indices)
train_indices, valid_indices = indices[:split], indices[split:]

# train 및 validation set의 index에 해당하는 데이터를 추출할 sampler 생성
train_sampler = torch.utils.data.sampler.SubsetRandomSampler(train_indices)
valid_sampler = torch.utils.data.sampler.SubsetRandomSampler(valid_indices)
```

```python
# Dataloader 구축
train_loader = torch.utils.data.DataLoader(trainset, batch_size=batch_size, num_workers=4, sampler=train_sampler)
valid_loader = torch.utils.data.DataLoader(trainset, batch_size=batch_size, num_workers=4, sampler=valid_sampler)
test_loader = torch.utils.data.DataLoader(testset, batch_size=batch_size, shuffle=False, num_workers=4)
```

# <br>__2. Model: GoogLeNet, ResNet, DenseNet__

### Step 2-1. Pretrained CNN Models 살펴보기
- Fully-connected layer 설명
    - **torch.nn.Linear(in_features, out_features, bias=True, device=None, dtype=None)**
    - in_features: input sample의 크기
    - out_features: output sample의 크기 (본 실습에서는 class의 개수)
    - bias: bias 추가 여부

### 1) GoogLeNet (Inception V3)
- Fine-tuning을 위해 auxiliary output을 도출하는 **(AuxLogits.fc)** layer와 primary output을 도출하는 **(fc)** layer를 변경해야 함<br>
- Inveption V3는 다른 pretrained CNN models와 다르게 auxiliary task를 수행하는 layer를 가지므로 해당 layer도 변경해야한다는 점을 주의해야 함

```python
# Pretrained Inveption V3 불러오기
inception = torchvision.models.inception_v3(pretrained=True)
print(inception)
```

```python
# (AuxLogits.fc) layer 및 (fc) layer 변경
inception.AuxLogits.fc = nn.Linear(768, num_classes)
inception.fc = nn.Linear(2048, num_classes)
print(inception)
```

### 2) ResNet
- Fine-tuning을 위해 output을 도출하는 **(fc)** layer를 변경해야 함

```python
# Pretrained Resnet18 불러오기
resnet = torchvision.models.resnet18(pretrained=True)
print(resnet)
```

```python
# (fc) layer 변경
resnet.fc = nn.Linear(512, num_classes)
print(resnet)
```

### 3) DenseNet
- Fine-tuning을 위해 output을 도출하는 **(classifier)** layer를 변경해야 함

```python
# Pretrained Densenet121 불러오기
densenet = torchvision.models.densenet121(pretrained=True)
print(densenet)
```

```python
# (classifier) layer 변경
densenet.classifier = nn.Linear(1024, num_classes)
print(densenet)
```

### Step 2-2. 모델 구축 및 학습 parameter 설정하기

```python
# feature extracting 여부에 따른 학습 parameter 설정
# feature_extracting=True일 때, the reshaped layer params만 학습하므로 이를 제외한 나머지 params는 gradient를 도출하지 않도록 설정함
def set_parameter_requires_grad(model, feature_extracting):
    if feature_extracting:
        for param in model.parameters():
            param.requires_grad = False  # True: gradient 전파하여 학습함, False: gradient 전파하지 않음
```

```python
# GoogLeNet, ResNet, DenseNet을 목적에 맞게 구축할 수 있는 함수
def initialize_model(model_name, num_classes, feature_extract, use_pretrained=True):
    model_ft = None

    if model_name == "resnet":
        """ Resnet18
        """
        # pretrained 모델 불러오기
        model_ft = models.resnet18(pretrained=use_pretrained)
        # 학습 parameter 설정
        set_parameter_requires_grad(model_ft, feature_extract)
        # fc layer 대체
        num_ftrs = model_ft.fc.in_features
        model_ft.fc = nn.Linear(num_ftrs, num_classes)

    elif model_name == "densenet":
        """ Densenet121
        """
        # pretrained 모델 불러오기
        model_ft = models.densenet121(pretrained=use_pretrained)
        # 학습 parameter 설정
        set_parameter_requires_grad(model_ft, feature_extract)
        # classifier layer 대체
        num_ftrs = model_ft.classifier.in_features
        model_ft.classifier = nn.Linear(num_ftrs, num_classes)

    elif model_name == "inception":
        """ Inception v3
        Be careful, expects (299,299) sized images and has auxiliary output
        """
        # pretrained 모델 불러오기
        model_ft = models.inception_v3(pretrained=use_pretrained)
        # 학습 parameter 설정
        set_parameter_requires_grad(model_ft, feature_extract)
        # auxilary network의 AuxLogits.fc layer 대체
        num_ftrs = model_ft.AuxLogits.fc.in_features
        model_ft.AuxLogits.fc = nn.Linear(num_ftrs, num_classes)
        # primary network의 fc layer 대체
        num_ftrs = model_ft.fc.in_features
        model_ft.fc = nn.Linear(num_ftrs,num_classes)

    return model_ft
```

```python
# model_name에 해당하는 initial 모델 구축
model_ft = initialize_model(model_name, num_classes, feature_extract, use_pretrained=True)
model_ft = model_ft.to(device)
print(model_ft)
```

# <br>__3. Training__

### Step 3-1. Optimizer 구축하기
- Finetuning: 모델의 모든 parameter를 업데이트함
- Feature extraction: 초기화한 layer의 parameter만 업데이트하고 나머지는 고정함

```python
# 업데이트 할 parameter 설정하기
params_to_update = model_ft.parameters()

print("Params to learn:")
if feature_extract:
    params_to_update = []
    # Feature extraction일 때, param.requires_grad == True인 parameter만 업데이트 할 parameter로 설정함
    for name, param in model_ft.named_parameters():
        if param.requires_grad == True:  
            params_to_update.append(param)
            print("-", name)
else:
    # Finetuning일 때, 모든 parameter를 업데이트 할 parameter로 설정함
    for name, param in model_ft.named_parameters():
        if param.requires_grad == True:  
            print("-", name)
```

```python
# params_to_update의 parameter만 업데이트 할 SGD optimizer 구축하기
optimizer_ft = optim.SGD(params_to_update, lr=0.001, momentum=0.9)
```

### Step 3-2. 모델 학습하기

```python
def train_model(model, dataloaders, criterion, num_epochs, optimizer, is_inception=False):
    since = time.time()

    val_acc_history = []

    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    for epoch in range(num_epochs):
        print('Epoch {}/{}'.format(epoch + 1, num_epochs))
        print('-' * 10)

        # 각 epoch마다 순서대로 training과 validation을 진행
        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # 모델을 training mode로 설정
            else:
                model.eval()   # 모델을 validation mode로 설정

            running_loss = 0.0
            running_corrects = 0
            running_total = 0

            # training과 validation 단계에 맞는 dataloader에 대하여 학습/검증 진행
            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)

                # parameter gradients를 0으로 설정
                optimizer.zero_grad()

                # forward
                # training 단계에서만 gradient 업데이트 수행
                with torch.set_grad_enabled(phase == 'train'):
                    # input을 model에 넣어 output을 도출한 후, loss를 계산함
                    # GoogLeNet을 학습할 경우 auxiliary task를 수행하므로 해당 output에 대한 loss를 추가적으로 계산함
                    if is_inception and phase == 'train':
                        outputs, aux_outputs = model(inputs)
                        loss1 = criterion(outputs, labels)
                        loss2 = criterion(aux_outputs, labels)
                        loss = loss1 + 0.4 * loss2
                    else:
                        outputs = model(inputs)
                        loss = criterion(outputs, labels)

                    # output 중 최댓값의 위치에 해당하는 class로 예측을 수행
                    _, preds = torch.max(outputs, 1)

                    # backward (optimize): training 단계에서만 수행
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # batch별 loss 축적
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)
                running_total += labels.size(0)

            # epoch의 loss 및 accuracy 도출
            epoch_loss = running_loss / running_total
            epoch_acc = running_corrects.double() / running_total

            print('{} Loss: {:.4f} Acc: {:.4f}'.format(phase, epoch_loss, epoch_acc))

            # validation 단계에서 validation loss가 감소할 때마다 best model 가중치를 업데이트함
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())
            if phase == 'val':
                val_acc_history.append(epoch_acc)

        print()

    # 전체 학습 시간 계산
    time_elapsed = time.time() - since
    print('Training complete in {:.0f}m {:.0f}s'.format(time_elapsed // 60, time_elapsed % 60))
    print('Best val Acc: {:4f}'.format(best_acc))

    # validation loss가 가장 낮았을 때의 best model 가중치를 불러와 best model을 구축함
    model.load_state_dict(best_model_wts)
    
    # best model 가중치 저장
    # torch.save(best_model_wts, '../output/best_model.pt')
    return model, val_acc_history
```

```python
# trining 단계에서 사용할 Dataloader dictionary 생성
dataloaders_dict = {
    'train': train_loader,
    'val': valid_loader
}
```

```python
# loss function 설정
criterion = nn.CrossEntropyLoss()
```

```python
# 모델 학습
model_ft, hist = train_model(model_ft, dataloaders_dict, criterion, num_epochs, optimizer_ft,
                             is_inception=(model_name=="inception"))
```

# <br>__4. Testing__

```python
def test_model(model, test_loader):
    model.eval()   # 모델을 validation mode로 설정
    
    # test_loader에 대하여 검증 진행 (gradient update 방지)
    with torch.no_grad():
        corrects = 0
        total = 0
        for inputs, labels in test_loader:
            inputs = inputs.to(device)
            labels = labels.to(device, dtype=torch.long)

            # forward
            # input을 model에 넣어 output을 도출
            outputs = model(inputs)

            # output 중 최댓값의 위치에 해당하는 class로 예측을 수행
            _, preds = torch.max(outputs, 1)

            # batch별 정답 개수를 축적함
            corrects += torch.sum(preds == labels.data)
            total += labels.size(0)

    # accuracy를 도출함
    test_acc = corrects.double() / total
    print('Testing Acc: {:.4f}'.format(test_acc))
```

```python
# 모델 검증 (Acc: 0.9020)
test_model(model_ft, test_loader)
```
